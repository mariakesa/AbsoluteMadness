```python
import wave
import numpy as np
import matplotlib.pyplot as plt

# Function to plot the waveform of a .wav file
def plot_waveform(wav_file):
    # Open the .wav file
    with wave.open(wav_file, 'r') as wav:
        # Extract Raw Audio from Wav File
        signal = wav.readframes(-1)
        signal = np.frombuffer(signal, dtype=np.int16)
        
        # Get the frame rate
        framerate = wav.getframerate()
        
        # Create a time array for plotting
        time = np.linspace(0, len(signal) / framerate, num=len(signal))
        
        # Plot the waveform
        plt.figure(figsize=(10, 4))
        plt.plot(time, signal)
        plt.title('Waveform of {}'.format(wav_file))
        plt.xlabel('Time [s]')
        plt.ylabel('Amplitude')
        plt.show()

# Replace 'yourfile.wav' with the path to your .wav file
plot_waveform('/home/maria/Music/Track 1.wav')
```
