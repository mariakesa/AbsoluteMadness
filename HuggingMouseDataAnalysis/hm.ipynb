```python
from sklearn.decomposition import PCA
from HuggingMouse.visualizers import VisualizerDimReduction
from HuggingMouse.trial_averaged_data import MakeTrialAveragedData
import numpy as np
allen_cache_path='/media/maria/DATA/AllenData'
boc = BrainObservatoryCache(manifest_file=str(
            Path(allen_cache_path) / 'brain_observatory_manifest.json'))
experiment_container = boc.get_experiment_containers()
rng = np.random.default_rng(78)
exp_ids=[dct['id'] for dct in experiment_container]
random_exp_ids = rng.choice(exp_ids, size=100, replace=False)
os.environ["HGMS_ALLEN_CACHE_PATH"]=allen_cache_path
dim_reduction_model = PCA(n_components=3)
id=random_exp_ids[2]
trial_averaged_data = MakeTrialAveragedData().get_data(id)
visualizer = VisualizerDimReduction(dim_reduction_model)
print(trial_averaged_data['three_session_A']['natural_movie_one'])
visualizer.info()
visualizer.visualize(trial_averaged_data,
                    'three_session_A', 'natural_movie_one')
```

```python
from sklearn.base import clone
def return_PCs(dim_reduction_model, trial_averaged_data):
    model = clone(dim_reduction_model)
    try:
        X_new = model.fit_transform(trial_averaged_data[str('three_session_A')][str('natural_movie_one')])
        return X_new
    except:
        print('boom')
counter=0
for id in random_exp_ids:
    print('My Counter!: ',counter)
    dim_reduction_model = PCA(n_components=3)
    trial_averaged_data = MakeTrialAveragedData().get_data(id)
    dat=return_PCs(dim_reduction_model, trial_averaged_data)
    try:
        print(dat.shape)
        if counter==0:
            my_data=np.array(dat)
        else:
            my_data=np.vstack((my_data,dat))
            print(my_data.shape)
    except:
        print('this one\'s out')
    counter+=1
np.save('my_dat.npy', my_data)
```
