```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load the time series data
with open("/media/maria/DATA/BrainObservatoryProcessedData/openai_clip-vit-base-patch32_embeddings.pkl", 'rb') as f:
    ts = pickle.load(f)['natural_movie_one'][:,10]


# Take the Fourier transform of ts
fft_result = np.fft.fft(ts)

# Calculate the frequency bins
n = len(ts)
timestep = 1  # assuming a unit time step
freq = np.fft.fftfreq(n, d=timestep)

# Plot the Fourier transform
plt.figure(figsize=(10, 6))
plt.plot(freq, np.abs(fft_result))
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of Time Series')
plt.grid(True)
plt.show()
```

```python
import pickle
import numpy as np
import matplotlib.pyplot as plt

# Load the time series data
with open("/media/maria/DATA/BrainObservatoryProcessedData/openai_clip-vit-base-patch32_embeddings.pkl", 'rb') as f:
    ts = pickle.load(f)['natural_movie_one'][:,0]

# Take the Fourier transform of ts
fft_result = np.fft.fft(ts)

# Calculate the frequency bins
n = len(ts)
timestep = 1  # assuming a unit time step
freq = np.fft.fftfreq(n, d=timestep)

# Plot the Fourier transform with logarithmic y scale
plt.figure(figsize=(10, 6))
plt.plot(freq, np.abs(fft_result))
plt.yscale('log')  # Set y-axis scale to logarithmic
plt.xlabel('Frequency (Hz)')
plt.ylabel('Amplitude')
plt.title('Fourier Transform of Time Series')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_cos_sin(t_start, t_end, num_points):
    t = np.linspace(t_start, t_end, num_points)
    y = np.cos(t) + np.sin(t)
    
    plt.figure(figsize=(8, 6))
    plt.plot(t, y)
    plt.xlabel('t')
    plt.ylabel('cos(t) + sin(t)')
    plt.title('Plot of cos(t) + sin(t)')
    plt.grid(True)
    plt.show()

# Example usage
plot_cos_sin(0, 2*np.pi, 1000)
```
